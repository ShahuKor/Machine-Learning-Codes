```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placements.csv')
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x=df['cgpa'],y=df['package'])
```

```python
X =df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
x_test
```

```python
lr.predict(x_test.iloc[1].values.reshape(1,1))
```

```python
y_test
```

```python
sns.scatterplot(data=df,x=df['cgpa'],y=df['package'])
plt.plot(x_test,lr.predict(x_test),color='red')
```

```python
# The red line represents the best fit line that our model has found after getting trained on the dataset
# As we know the equation of line is y= mx + c' so, Basically Our model found out the best Slope(m) and the best Y intercept(c')
```

```python
#Lets the what those values are
```

```python
m = lr.coef_
```

```python
m
```

```python
c = lr.intercept_
```

```python
c
```

```python
# Behind the scenes of the prediction our model is basically solving this equation:
# y = mx + c

#Lets take an Example :
x= 7.15 #Any random Value of X_test

y = m * x + c
```
