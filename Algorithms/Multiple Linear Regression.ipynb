```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=100,n_features=2,n_targets=1,n_informative=2,noise=50)
```

```python
X
```

```python
Y
```

```python
df = pd.DataFrame({'Fe1':X[:,0],'Fe2':X[:,1],'Target':Y})
```

```python
df.head()
```

```python
# We generated random data from make_regression class which is specifically made to make random data for regression problems
```

```python
# Lets plot our 3D data into a scatter plot in 3D for better visualisation
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df,x='Fe1',y='Fe2',z='Target')
```

```python
fig.show()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, r2_score, mean_absolute_error, mean_squared_error
```

```python
print("R2 Score : ", r2_score(y_test,y_pred))
print("MAE Score : ", mean_absolute_error(y_test,y_pred))
print("MSE Score : ", mean_squared_error(y_test,y_pred))
```

```python
import plotly.graph_objects as go
```

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
xGrid, yGrid = np.meshgrid(y,x)


final = np.vstack((xGrid.ravel().reshape(1,100),yGrid.ravel().reshape(1,100))).T
z_final = lr.predict(final).reshape(10,10)

z = z_final

```

```python
fig = px.scatter_3d(df, x='Fe1', y='Fe2', z='Target')
fig.add_trace(go.Surface(x=x,y=y,z=z))
fig.show()
```
